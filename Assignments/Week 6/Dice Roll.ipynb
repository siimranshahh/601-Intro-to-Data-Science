```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Getting the probablities for the dice roll: 
def dice_rolling(num_rolls):
    fair_dice = [1/6]*6
    biased_dice = [0.15, 0.15, 0.15, 0.15, 0.15, 0.25]
    
    #The dice roll for fair and biased dice: 
    fair_rolls = np.random.choice(range(1, 7), size=num_rolls, p=fair_dice) #the probablity is 1/6th that of the fair dice and using random choice
    biased_rolls = np.random.choice(range(1, 7), size=num_rolls, p=biased_dice) #the probablity is that of the biased dice and using random choice

    return fair_rolls, biased_rolls
```

```python
num_rolls = 1000
fair_rolls, biased_rolls = dice_rolling(num_rolls)
```

### Plotting historgram for outcomes of the Fair Dice

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(fair_rolls, bins=np.arange(1, 8), alpha=0.7, color='blue', rwidth=0.85)
plt.title('Fair Die Rolls')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.xticks(range(1, 7))

plt.tight_layout()
plt.show()
```

### Plotting historgram for outcomes of the Biased Dice

```python
plt.subplot(1, 2, 2)
plt.hist(biased_rolls, bins=np.arange(1, 8), alpha=0.7, color='red', rwidth=0.85)
plt.title('Biased Die Rolls')
plt.xlabel('Outcome')
plt.xticks(range(1, 7))

plt.tight_layout()
plt.show()
```

#### From the above historgrams, it is understood that in the fair dice all numbers have almost the same probability. In the Biased dice the probablity of numbers 1-5 is almost similar but 6 number outcome is higher.
