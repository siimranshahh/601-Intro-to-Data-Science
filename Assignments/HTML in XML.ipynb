## HTML in XML
I provide you with a compressed XML file. Some of the fields contain HTML. ​

1) Extract the XML from the .zip file. In Python, use a module to parse the XML (do not write an XML parser)​
   
2) Using Python, extract the HTML from the XML. Then use BeautifulSoup4 to parse the HTML
   
3) For each HTML page, report the number of links (URLs with the tag < a href="URL">text) in each HTML page
   
4) Submit a single Jupyter notebook that parses the XML file and produces the count of links per HTML file.​

```python
import zipfile
import pandas as pd

from bs4 import BeautifulSoup
```

Getting zip file and xml file and saving them in `Zip_path` and `Xml_path`

```python
zip_path = 'xml_containing_html.xml.zip'
xml_path = 'xml_containing_html.xml'
```

Extracting the XML file from the zip using module zipfile

```python
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extract('xml_containing_html.xml', '.')    
```

Using pandas to read the XML file

```python
df = pd.read_xml(xml_path)
```

```python
df.head
```

### Using Beautiful Soup to parse html file

 - final_all('a') to find all anchor tags in the html file
 - returns number of tags found (len(links))

```python
def links_html(html):
    soup = BeautifulSoup(html, 'html.parser')
    links = soup.find_all('a')
    return len(links)
```

`num_links` = number of links on each page 

```python
df['num_links'] = df['content'].apply(links_html)
```
